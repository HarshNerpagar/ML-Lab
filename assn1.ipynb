```python
data_series = pd.Series(df1['Age']);

# Calculate Q1 and Q3
q1 = data_series.quantile(0.25)
q3 = data_series.quantile(0.75)

# Calculate the IQR
iqr = q3 - q1
lb = q1-1.5*iqr
ub = q3+1.5*iqr
print(iqr,lb,ub)
```

```python
columns_to_plot = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
plt.figure(figsize=(10, 6))
plt.boxplot([df[col] for col in columns_to_plot], labels=columns_to_plot)
plt.title("Boxplot of Selected Columns")
plt.ylabel('Values')
plt.show()
```

```python
df_cleaned = df[(df['Age'] >= lb) & (df['Age'] <= ub)]
df_cleaned
```

```python
columns_to_plot = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Plot multiple boxplots
plt.figure(figsize=(12, 10))
plt.boxplot([df_cleaned[col] for col in columns_to_plot], labels=columns_to_plot)
plt.title("Boxplot of Selected Columns")
plt.ylabel('Values')
plt.show()
```

```python
sns.countplot(x='Survived', data=df) # Changed 'survived' to 'Survived'
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df) # Changed 'sex' to 'Sex' and 'survived' to 'Survived'
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df) # Changed 'pclass' to 'Pclass' and 'survived' to 'Survived'
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

```python
# Convert categorical data to numeric if needed
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Create correlation matrix
corr = df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
